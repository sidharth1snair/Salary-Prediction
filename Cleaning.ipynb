```python
import pandas as pd
import re

df = pd.read_csv("glassdoor_jobs_final_data.csv")  # Replace with your actual file path

# Remove rows where 'Salary Estimate' is 'N/A'
df_cleaned = df[df['Salary Estimate'].notna()]  # Removes NaN
df_cleaned = df_cleaned[df_cleaned['Salary Estimate'] != 'N/A']  # Removes literal 'N/A'

# Optionally, reset the index
df_cleaned.reset_index(drop=True, inplace=True)

```

```python
df_cleaned
```

```python
def extract_numeric_salary(text):
    text = text.lower()
    # Match salary ranges
    match_range = re.search(r'₹?(\d+)[lk]?\s*–\s*₹?(\d+)[lk]?/([a-z]+)', text)
    if match_range:
        return f"{match_range.group(1)} – {match_range.group(2)}/{match_range.group(3)}"
    # Match single salary value
    match_single = re.search(r'₹?(\d+)[lk]?/([a-z]+)', text)
    if match_single:
        return f"{match_single.group(1)}/{match_single.group(2)}"
    return None

# Apply the function to the column
df_cleaned['Cleaned Salary'] = df_cleaned['Salary Estimate'].apply(extract_numeric_salary)

```

```python
df_cleaned
```

```python
def is_monthly(salary_str):
    if salary_str and '/mo' in salary_str:
        return 1
    return 0

df_cleaned['monthly'] = df_cleaned['Cleaned Salary'].apply(is_monthly)
```

```python
df_cleaned.head()
```

```python
def extract_min_max(s):
    if pd.isna(s):
        return pd.Series([None, None])
    try:
        parts = s.split('/')[0].split('–')
        min_salary = int(parts[0].strip())
        max_salary = int(parts[1].strip()) if len(parts) > 1 else int(parts[0].strip())
        return pd.Series([min_salary, max_salary])
    except:
        return pd.Series([None, None])

df_cleaned[['min_salary', 'max_salary']] = df_cleaned['Cleaned Salary'].apply(extract_min_max)
```

```python
df_cleaned.head()
```

```python
def convert_to_rupees(row):
    factor = 1000 if row['monthly'] == 1 else 100000
    min_r = row['min_salary'] * factor
    max_r = row['max_salary'] * factor
    return pd.Series([min_r, max_r])

df_cleaned[['min_salary_rupees', 'max_salary_rupees']] = df_cleaned.apply(convert_to_rupees, axis=1)
df_cleaned['avg_salary_rupees'] = (df_cleaned['min_salary_rupees'] + df_cleaned['max_salary_rupees']) / 2
```

```python
# Clean 'Company Name' by removing newline and anything after it
df_cleaned['company_name_cleaned'] = df_cleaned['Company Name'].apply(lambda x: str(x).split('\n')[0].strip())
df_cleaned = df_cleaned[
    (df_cleaned['company_name_cleaned'].notna()) & 
    (df_cleaned['company_name_cleaned'].str.lower().str.strip() != 'nan') & 
    (df_cleaned['company_name_cleaned'].str.strip() != "")
]

```

```python
# Get unique job titles (remove NaN, strip whitespaces)
unique_titles = df_cleaned['Job Title'].dropna().str.strip().unique()

# Convert to a sorted list
unique_titles = sorted(unique_titles)

# Display them
for title in unique_titles:
    print(title)
```

```python
def map_job_title(title):
    title = title.lower().strip()

    # Existing mappings
    if any(x in title for x in ['data scientist', 'decision scientist', 'applied scientist', 'statistical data']):
        return 'Data Scientist'
    elif any(x in title for x in ['ml engineer', 'machine learning engineer', 'ml developer', 'ml/ai developer']):
        return 'ML Engineer'
    elif any(x in title for x in ['ai', 'artificial intelligence', 'agentic ai', 'generative ai', 'gen ai', 'aiml']):
        return 'AI Engineer'
    elif any(x in title for x in ['data analyst', 'business analyst', 'bi analyst', 'marketing analyst']):
        return 'Data Analyst'
    elif 'engineer' in title and 'data' in title:
        return 'Data Engineer'
    elif any(x in title for x in ['product manager', 'program manager', 'project manager']):
        return 'Product/Program Manager'
    elif any(x in title for x in ['manager', 'director', 'vp', 'lead', 'head']):
        return 'Manager/Director'
    elif any(x in title for x in ['trainer', 'faculty', 'mentor', 'tutor']):
        return 'Trainer/Faculty'
    elif any(x in title for x in ['intern', 'trainee', 'junior', 'associate']):
        return 'Intern/Associate'

    # New enhanced mappings from unique titles
    elif 'full stack' in title:
        return 'Full Stack Developer'
    elif 'backend' in title or ('developer' in title and 'python' in title and 'frontend' not in title and 'full' not in title):
        return 'Backend Developer'
    elif 'frontend' in title or ('front end' in title) or ('react' in title and 'backend' not in title):
        return 'Frontend Developer'
    elif 'web developer' in title or 'wordpress' in title:
        return 'Web Developer'
    elif 'software developer' in title or 'software engineer' in title:
        return 'Software Engineer'
    elif 'python developer' in title and 'ai' not in title and 'ml' not in title:
        return 'Python Developer'
    elif 'java developer' in title:
        return 'Java Developer'
    elif 'react developer' in title:
        return 'React Developer'
    elif 'nodejs' in title or 'node js' in title:
        return 'NodeJS Developer'
    elif 'unity' in title:
        return 'Game Developer'
    elif 'game developer' in title:
        return 'Game Developer'
    elif 'ui developer' in title or 'ui path developer' in title or 'uipath' in title:
        return 'UI Developer'
    elif 'rpa' in title:
        return 'RPA Developer'
    elif 'cloud' in title:
        return 'Cloud Engineer'
    elif 'salesforce' in title:
        return 'Salesforce Developer'
    elif 'power bi' in title or 'bi developer' in title:
        return 'BI Developer'
    elif 'support engineer' in title or 'technical support' in title:
        return 'Support Engineer'
    elif 'shiny' in title or 'sql developer' in title:
        return 'SQL Developer'
    elif 'databricks' in title:
        return 'Data Engineer'
    elif 'django' in title:
        return 'Python Developer'
    elif 'html' in title or 'css' in title or 'tailwind' in title:
        return 'Frontend Developer'
    elif 'dashboard' in title or 'reporting' in title:
        return 'BI Developer'
    
    return 'Other'


df_cleaned['Standardized Title'] = df_cleaned['Job Title'].apply(map_job_title)
df_cleaned['Standardized Title'].value_counts()

```
